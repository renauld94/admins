## 1. Accessing Healthcare Data via APIs

```python
# Example: Fetching healthcare data from a public API
import requests
response = requests.get('https://api.publicapis.org/entries')
if response.status_code == 200:
    data = response.json()
    print('Number of APIs listed:', len(data['entries']))
else:
    print('Failed to fetch data')
```

## 2. Parsing JSON Data

```python
# Example: Parsing patient data from JSON
import json
patient_json = '{"id": "PT001", "name": "John Doe", "age": 45, "diagnosis": "Hypertension"}'
patient = json.loads(patient_json)
print('Patient name:', patient['name'])
```
