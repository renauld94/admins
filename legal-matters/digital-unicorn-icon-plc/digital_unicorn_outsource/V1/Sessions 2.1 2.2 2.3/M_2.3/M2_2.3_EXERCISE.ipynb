## Exercise D ##

üìù Task:

Use a while loop with variable `LBORRES` set equal to zero initially (`LBORRES = 0`) and increments by one each time under the condition `LBORRES < 5` is not satisfied.

‚úÖ Solution

```python
LBORRES = 0
while LBORRES < 5:
    LBORRES += 1
    print("New value:", LBORRES)
```

## Exercise E ##
Nesting example

üìù Task:
* If string variable `AETERM = headache` then print out `headache` three times.
* Else print out `Unknown AETERM`

‚úÖ Solution

```python
AETERM = "headache"

if AETERM == "headache":
    for i in range(3):
        print("headache")
else:
    print("Unknown AETERM")
```
