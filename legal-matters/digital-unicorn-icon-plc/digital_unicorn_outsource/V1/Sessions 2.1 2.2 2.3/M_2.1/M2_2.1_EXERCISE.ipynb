## Exercise E ##

Lists can consist of elements of different types, for instance strings, floats and integers could all be within the same list.

üìù Task:

1. Print the value of the third element of the list `AETERM = ["Headache", "Back pain", "Cough", "Fever", "Nausea", "Vomiting", "Diarrhea"]`
2. Print the last element of the list 
3. Print the list reversed in order
4. Append an additional fruit (e.g. "Abdominal pain") to the list and print the result
5. Append an integer to the list and then print

‚úÖ Solution


```python
AETERM = ["Headache", "Back pain", "Cough", "Fever", "Nausea", "Vomiting", "Diarrhea"]

# print the third element in the list
print(AETERM[2])

# print the last element in the list
print(AETERM[-1])

# flip the list and reverse the order
print(AETERM[::-1])

# append an extra fruit of your choice
AETERM.append("Abdominal pain")
print(AETERM)

# append another data type
AETERM.append(1)
print(AETERM)
```
