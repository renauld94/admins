## Variables and Data Types ##

### Integers and Floats ###
* Integers are whole numbers, e.g. `7`
* Floats are numbers with decimals e.g. `6.71`

### Example ###

```python
# Store dimensions
WIDTH = 20 # in millimeters
HEIGHT = 15 # in millimeters

# Calculate area
AREA = WIDTH * HEIGHT

# Display result
print("Area:", AREA, "sq mm")
# Output: Area: 300 sq mm
```

### Strings and Text ###

* String or Text are also data types just like integer and float.
* Integers and float are numbers but string accepts anything within quotes.
* Text enclosed in single (`'text'`) or double quotes (`"text"`)
* Combine strings with + operator like `"Hello"+"World"` will be `"HelloWorld"`
* Repeat strings with  operator like `"s"*3` will show `sss`

## Example ##

```python
# Combine strings
GREETING = 'Hello ' + 'World'
print(GREETING)  # Hello World

# Repeat strings
print('un' * 3 + 'ique')  # unununique

# Clinical application
PATIENT = "J. Smith"
MESSAGE = "Patient " + PATIENT + " scheduled"
print(MESSAGE)
```

### Lists and Collections ###
* Lists group multiple values in square brackets `[ ]`
* Can contain different data types
* Index and slice like strings: `list[0]`, `list[1:3]`
* Mutable: can change elements or add new ones
* Add elements with `.append()` method
* Get length with `len()` function
* Lists can contain other lists for complex nested structures

## Example ##

```python
# Create a list of squares
SQUARES = [1, 4, 9, 16, 25]
print(SQUARES[0])     # 1 (first element)
print(SQUARES[-1])    # 25 (last element)

# Add new elements
SQUARES.append(36)
print(SQUARES)        # [1, 4, 9, 16, 25, 36]

# List of patient data
PATIENT_DATA = [
    "Smith", 45, "Male", [120, 80], True
]   # Name, age, gender, BP, active status
print("Patient age:", PATIENT_DATA[1])
```
