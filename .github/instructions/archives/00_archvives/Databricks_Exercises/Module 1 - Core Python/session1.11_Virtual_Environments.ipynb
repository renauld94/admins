## 1. Creating a Virtual Environment

# Create a virtual environment
# Run in terminal: python3 -m venv healthcare_env
# Activate: source healthcare_env/bin/activate
# Deactivate: deactivate
print('Virtual environment commands:')
print('python3 -m venv healthcare_env')
print('source healthcare_env/bin/activate')
print('deactivate')
